```python
p = Path("./images")
data_dir = p
print(data_dir)
print(type(data_dir))
bilder_liste = list(p.glob("*.jpg"))
image_count = len(bilder_liste)
print(image_count)
for bild in range(len(bilder_liste)):
    zahl = int(str(bilder_liste[bild])[-7:-4])
    print(zahl)
    if zahl <= 59:
        shutil.move(str(bilder_liste[bild]), './images/A045-59')
    elif zahl <= 74:
        shutil.move(str(bilder_liste[bild]), './images/A060-74')
    elif zahl <= 86:
        shutil.move(str(bilder_liste[bild]), './images/A075-86')
    elif zahl <= 93:
        shutil.move(str(bilder_liste[bild]), './images/A087-93')
    elif zahl <= 104:
        shutil.move(str(bilder_liste[bild]), './images/A094-104')
    elif zahl <= 119:
        shutil.move(str(bilder_liste[bild]), './images/A105-119')
    else:
        shutil.move(str(bilder_liste[bild]), './images/A120-135')
```
